```python
import pandas as pd
import numpy as np
import seaborn as sbn
import matplotlib.pyplot as plt
```

# Importando dados de Imigracao para o Canada


```python
df0 = pd.read_csv('/content/imigrantes_canada.csv')
df0
```

```python
df0.info()
```

# Transformando nome dos Paises em Index

```python
df0 = df0.set_index('País')
```

```python
df0
```

```python
anos = list(map(str, range(1980, 2014)))
```

```python
anos
```

# Imigrantes Brasil para o Canada


```python
brasil = df0.loc['Brasil', anos]
```

```python
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantesBrasil': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil
```

```python
plt.figure(figsize=(10, 4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantesBrasil'])
plt.legend(['Brasil'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010', '2014'])
plt.yticks([500, 1000, 1500, 2000, 2500, 3000])
plt.title('Imigrantes Brasileiros para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Quantidade de Imigrantes')
plt.show()
```

# Codigo realizado por mim para comparar os imigrantes entre Brasil e Argentina

```python
brasil = df0.loc['Brasil', anos]
argentina = df0.loc['Argentina', anos]
```

```python
brasil, argentina
```

```python
sulamerica_dict = {'ano': brasil.index.tolist(), 'imigrantesBrasil': brasil.values.tolist(), 'imigrantesArgentina': argentina.values.tolist()}
dados_sulamerica = pd.DataFrame(sulamerica_dict)
```

```python
dados_sulamerica
```

```python
plt.figure(figsize=(10, 4))
plt.plot(dados_sulamerica['ano'], dados_sulamerica['imigrantesBrasil'], label='Brasil')
plt.plot(dados_sulamerica['ano'], dados_sulamerica['imigrantesArgentina'], label='Argentina')
plt.legend(['Brasil', 'Argentina'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010', '2014'])
plt.yticks([500, 1000, 1500, 2000, 2500, 3000])
plt.title('Imigrantes Sulamericanos para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Quantidade de Imigrantes')
plt.show()
```

```python
sulamerica = df0.query('Região == "América do Sul"')
sulamerica
```

```python
cores = ['r', 'g', 'b', 'y', 'c', 'm', 'k']
fig, ax = plt.subplots(figsize=(12,5))
ax.bar(sulamerica.index, sulamerica['Total'], color=cores)
ax.set_title('Imigrantes  Sulamericanos para o Canada\n(1980 a 2013)', fontsize=18, y=1.02, loc='left')
ax.set_xlabel('')
ax.set_ylabel('Numero Total de Imigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.show
```

```python
cores = ['r', 'g', 'b', 'y', 'c', 'm', 'k']
fig, ax = plt.subplots(figsize=(12,5))
ax.barh(sulamerica.index, sulamerica['Total'], color=cores)
ax.set_title('Imigrantes  Sulamericanos para o Canada\n(1980 a 2013)', fontsize=18, y=1.02, loc='left')
ax.set_xlabel('Numero Total de Imigrantes', fontsize=14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.show
```

```python
sulamerica_sorted = sulamerica.sort_values('Total', ascending=True)
```

```python
cores = ['r', 'g', 'b', 'y', 'c', 'm', 'k']
fig, ax = plt.subplots(figsize=(12,5))
ax.barh(sulamerica_sorted.index, sulamerica_sorted['Total'], color=cores)
ax.set_title('Imigrantes  Sulamericanos para o Canada\n(1980 a 2013)', fontsize=18, y=1.02, loc='left')
ax.set_xlabel('Numero Total de Imigrantes', fontsize=14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.show
```

```python
cores = []
for pais in sulamerica_sorted.index:
  if pais == 'Brasil':
    cores.append('green')
  else :
    cores.append('silver')
fig, ax = plt.subplots(figsize=(12,5))
ax.barh(sulamerica_sorted.index, sulamerica_sorted['Total'], color=cores)
ax.set_title('America do Sul: Brasil foi o 4 pais com mais imigrantes\npara o Canada entre 1980 e 2013)', fontsize=18, y=1.02, loc='left')
ax.set_xlabel('Numero Total de Imigrantes', fontsize=14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.show
```
