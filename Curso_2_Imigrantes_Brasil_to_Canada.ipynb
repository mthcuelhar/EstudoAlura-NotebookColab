```python
import pandas as pd
import numpy as np
import seaborn as sbn
import matplotlib.pyplot as plt
```

# Importando dados de Imigracao para o Canada


```python
df0 = pd.read_csv('/content/imigrantes_canada.csv')
df0
```

```python
df0.info()
```

# Transformando nome dos Paises em Index

```python
df0 = df0.set_index('País')
```

```python
df0
```

```python
anos = list(map(str, range(1980, 2014)))
```

```python
anos
```

# Imigrantes Brasil para o Canada


```python
brasil = df0.loc['Brasil', anos]
```

```python
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantesBrasil': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil
```

```python
plt.figure(figsize=(10, 4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantesBrasil'])
plt.legend(['Brasil'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010', '2014'])
plt.yticks([500, 1000, 1500, 2000, 2500, 3000])
plt.title('Imigrantes Brasileiros para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Quantidade de Imigrantes')
plt.show()
```

# Codigo realizado por mim para comparar os imigrantes entre Brasil e Argentina

```python
brasil = df0.loc['Brasil', anos]
argentina = df0.loc['Argentina', anos]
```

```python
brasil, argentina
```

```python
sulamerica_dict = {'ano': brasil.index.tolist(), 'imigrantesBrasil': brasil.values.tolist(), 'imigrantesArgentina': argentina.values.tolist()}
dados_sulamerica = pd.DataFrame(sulamerica_dict)
```

```python
dados_sulamerica
```

```python
plt.figure(figsize=(10, 4))
plt.plot(dados_sulamerica['ano'], dados_sulamerica['imigrantesBrasil'], label='Brasil')
plt.plot(dados_sulamerica['ano'], dados_sulamerica['imigrantesArgentina'], label='Argentina')
plt.legend(['Brasil', 'Argentina'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010', '2014'])
plt.yticks([500, 1000, 1500, 2000, 2500, 3000])
plt.title('Imigrantes Sulamericanos para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Quantidade de Imigrantes')
plt.show()
```

# Codigo realizado pela Alura para comparar os imigrantes entre Brasil e Argentina

```python
df_comparacao = df0.loc[['Brasil', 'Argentina'], anos]
```

```python
df_comparacao = df_comparacao.T
df_comparacao.head()
```

```python
plt.plot(df_comparacao['Brasil'],label = 'Brasil')
plt.plot(df_comparacao['Argentina'],label ='Argentina')
plt.title('Imigração do Brasil e Argentina para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.legend()
plt.show()
```

```python
fig, axs = plt.subplots(2,2, figsize=(15,6))
fig.subplots_adjust(hspace=1)
fig.suptitle('Imigracao dos 4 maiores paises Sulmericanos para o Canada (1980 a 2013)', y=1.02)

axs[0,0].plot(df0.loc['Brasil',anos]),
axs[0,0].set_title('Brasil')

axs[0,1].plot(df0.loc['Argentina',anos])
axs[0,1].set_title('Argentina')

axs[1,0].plot(df0.loc['Peru',anos])
axs[1,0].set_title('Peru')

axs[1,1].plot(df0.loc['Colômbia',anos])
axs[1,1].set_title('Colômbia)')

for ax in axs.flat:
  ax.set(xlabel='Ano',
         ylabel='Numero de Imigrantes')
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.grid(True)

ymin = 0
ymax = 7000
for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)

plt.show()
```
