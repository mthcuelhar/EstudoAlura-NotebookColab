# Codigo realizado por mim para comparar os imigrantes entre Brasil e Argentina

```python
brasil = df0.loc['Brasil', anos]
argentina = df0.loc['Argentina', anos]
```

```python
brasil, argentina
```

```python
sulamerica_dict = {'ano': brasil.index.tolist(), 'imigrantesBrasil': brasil.values.tolist(), 'imigrantesArgentina': argentina.values.tolist()}
dados_sulamerica = pd.DataFrame(sulamerica_dict)
```

```python
dados_sulamerica
```

```python
plt.figure(figsize=(10, 4))
plt.plot(dados_sulamerica['ano'], dados_sulamerica['imigrantesBrasil'], label='Brasil')
plt.plot(dados_sulamerica['ano'], dados_sulamerica['imigrantesArgentina'], label='Argentina')
plt.legend(['Brasil', 'Argentina'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010', '2014'])
plt.yticks([500, 1000, 1500, 2000, 2500, 3000])
plt.title('Imigrantes Sulamericanos para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Quantidade de Imigrantes')
plt.show()
```

# Codigo realizado pela Alura para comparar os imigrantes entre Brasil e Argentina

```python
df_comparacao = df_comparacao.T
df_comparacao.head()
```

```python
plt.plot(df_comparacao['Brasil'],label = 'Brasil')
plt.plot(df_comparacao['Argentina'],label ='Argentina')
plt.title('Imigração do Brasil e Argentina para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.legend()
plt.show()
```
